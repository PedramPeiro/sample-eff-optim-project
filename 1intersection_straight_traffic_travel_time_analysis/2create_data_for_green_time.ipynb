For every SUMO simulation, we vary the green time on the 1 intersection light, therefore the inner loop for the green_time. And in addition to that, we can vary the number of vehicles with the different route files created in the last notebook, therefore the outer loop. Here, the number of vehicle is set to 1000.

For every SUMO simulation, three input files are needed:
the .net.xml (network file)
the .sumo.cfg (config file)
the .additional.xml file (for the edge probe)
and the .rou.xml file (the route file, generated in step 1)

each of these files are written in this notebook with the .net.xml file containing the green times information
and the SUMO simulation is run with the output files for further analysis.
